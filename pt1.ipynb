# TEORIA

# 1) Entidad de DES vs. Agente de MBA (ABM)

### Diferencia esencial (en palabras simples)

**DES (eventos discretos):** una entidad es algo que fluye por un proceso (llega → espera en cola → usa un recurso → sale).  
El foco está en el flujo, los recursos y los eventos (no “piensan”: reaccionan a eventos).

**MBA/ABM (agentes):** un agente es un individuo con reglas de comportamiento, capaz de decidir e interactuar con otros y con el entorno.  
El foco está en cómo múltiples decisiones locales producen patrones globales.

### ¿Por qué cada paradigma sirve para problemas diferentes?

- **DES:** ideal cuando lo central es capacidad–colas–tiempos en procesos bien definidos (p. ej., cuántos servidores/puestos necesito para cumplir un SLA).  
- **MBA:** ideal cuando interesan interacciones entre actores autónomos y heterogeneidad (p. ej., difusión de conductas o enfermedades en una población).

### Ejemplos breves

- **DES:** dimensionamiento de un centro de llamadas (puestos, horarios, tiempos de espera).  
- **MBA:** propagación de un comportamiento/enfermedad en una ciudad con individuos heterogéneos.

# 2) Por qué aparecen colas en DES y qué aprendemos de ellas

### Colas casi inevitables

Cuando la tasa de llegada se acerca o supera la tasa de servicio, se acumulan entidades esperando.  
Incluso con capacidad “suficiente”, la variabilidad (rachas de llegadas, tiempos de servicio irregulares) genera picos y, por tanto, colas.  
La **Ley de Little** resume el vínculo entre cantidad en cola, tasa de llegada y tiempo de espera:

\[
L = \lambda \times W
\]

### Información valiosa al analizar colas (a lo largo de la simulación)

- **Longitud promedio y percentiles de la cola →** decisiones de capacidad y SLA.  
- **Tiempo de espera promedio y percentiles →** calidad de servicio percibida.  
- **Variabilidad y picos por franja →** diseño de turnos, buffers y políticas de prioridad.  
- **Efecto de políticas** (colas unificadas vs. separadas, prioridades) → impacto en esperas y utilización.

# 3) Escenario con recursos móviles, estáticos y portátiles (no hospital)

### Aeropuerto mediano (terminal + plataforma)

- **Estáticos:** mostradores de check-in, arcos de seguridad, puertas de embarque (fijos; capacidad por puesto).  
- **Móviles:** autobuses de plataforma, remolcadores, escaleras móviles (se desplazan a distintas aeronaves).  
- **Portátiles:** tablets/escáneres para embarque, radios/PDA de supervisores (se llevan encima, amplían control/registro).

**Cómo interactúan:**  
Los pasajeros (entidades DES) fluyen por recursos estáticos (check-in, seguridad), mientras recursos móviles atienden aeronaves en posiciones variables y recursos portátiles permiten registrar/autorizar procesos sin depender de un puesto fijo.  
Este mix es típico en enfoques multimétodo.

# 4) Combinar MBA y DES: brote + vacunación

### Planteamiento
Se modela la propagación de una enfermedad (**MBA**) y se integra el proceso de vacunación en centros de salud (**DES**).

### Parte MBA (ABM)

Agentes-persona con reglas de movilidad, contactos, probabilidad de contagio, estado (S/I/R), aceptación de vacuna.  
**Salida clave:** incidencia por zona/semana, demanda potencial de vacunación, grupos prioritarios.

### Parte DES

Centros de salud como procesos con llegadas (desde MBA), colas, tiempos de registro/vacunación/recuperación y recursos (puestos, personal, cabinas, cadena de frío).  
**Salida clave:** tiempos de espera, longitud de cola, capacidad diaria, consumo de insumos.

### Información que se pasa entre modelos

- **MBA → DES:** tasa de llegadas por barrio/franja, perfiles de prioridad y adhesión.  
- **DES → MBA:** tasa de servicio efectiva (vacunados/día), retrasos (espera) que influyen en adhesión y en el ritmo de inmunización.

**Ventaja:** el multimétodo usa lo mejor de ambos: interacciones individuales (MBA) + gestión operativa de colas/recursos (DES).
